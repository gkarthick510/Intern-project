```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#plt.style.use('fivethirtyeight')
```

```python
file=pd.read_csv("/content/ICICIBANK.NS.csv")
file.head()
```

```python
file.shape
```

```python
file.isnull().any()
```

```python
file.info()
```

```python
file.describe()
```

```python
sns.heatmap(file.corr(),annot=True)
```

```python
sns.histplot(data=file,x="Close")
plt.title("ICICI STOCK PRICE DATA",fontsize=10)
```

```python
sns.displot(data=file,x="High")
```

```python
sns.jointplot(data=file,x="Close",y="High",kind="hist")
```

```python
file.Close.plot(kind="kde")
```

```python
plt.figure(figsize=(30,6))
plt.grid(True)
plt.xlabel('Date')
plt.ylabel('Close Prices')
sns.lineplot(data=file,x='Date',y='Close')
plt.title('ICICI BANK closing price')
plt.xticks(rotation='vertical')
plt.show()
```
